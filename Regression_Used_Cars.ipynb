# Regression Model for Used Cars

## Checking OLS assumptions
 1. linearity
 2. endogeneity 
 3. normality and homoscedasticity
 4. autocorrelation
 5. multicollinearity

#### checking for linearity using scatter plot

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'], data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'], data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'], data_cleaned['Price'])
ax3.set_title('Price and Mileage')

plt.show()
```

### Relaxing the assumptions

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'], data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'], data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'], data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')

plt.show()
```

#### linear patterns can be seen in the resulting plots above

```python
# we can drop the Price col as its no longer needed
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

### Multicollinearity
to check for this, we use VIF (variance inflation factor) from statsmodels

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

#### when vif = 1 => there is no multicollinearity
1 < vif < 5 => considered okay
some consider 10 to be the boundary, in this case Year looks to correlated and will be removed

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

```python
data_no_multicollinearity
```

## Create dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

### Rearrange Columns

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

### Linear Regression Model

#### declare the inputs

```python
targets = data_preprocessed.log_price
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

#### scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

#### train test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

#### create the regression

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.displot(y_train - y_hat)
plt.title('Residuals PDF', size=18)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# using summary table to display wieghts

reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
plt.scatter(y_train, y_hat, alpha=0.2)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
data_performance = pd.DataFrame(y_hat_test, columns=['Predictions'])
data_performance.head()
```

#### the above results are shown in log price, to invert back to normal prices:

```python
data_performance = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
data_performance.head()
```

```python
# placing the targets along with predictions for visual comparison
data_performance['Target'] = np.exp(y_test)
data_performance.head()
```

#### reason for the strange results is cos indexing is altered, to correct this:

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
data_performance.Target = np.exp(y_test)
data_performance.head()
```

#### to look at the difference btwn targets and predictions

```python
data_performance['Residual'] = data_performance.Target - data_performance.Predictions
```

```python
data_performance['%Difference'] = np.absolute(data_performance['Residual']/data_performance['Target']*100)
data_performance
```

```python
data_performance.describe()
```

```python
data_performance.sort_values(by=['%Difference'])
```
